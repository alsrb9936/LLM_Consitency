# Gemini

```python
import json
import os
import re
import pandas as pd
import os
import json
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


results_path = "../results"
api_type = "gemini"
data_name = "qnli"  # Example data name
prompt_num = ["2"]  # Example prompt number

root_path = os.path.join(results_path, api_type, data_name)

file_path = f"../data/raw/{data_name}.jsonl"
data = []
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        data.append(json.loads(line.strip()))

# convert data_list to DataFrame
df = pd.DataFrame(data)

y_true = df["label"].tolist()
total_results = []
for num in prompt_num:
    results_path_list = os.path.join(root_path, f"prompt_{num}", "output")
    tmp_list = []

    for n, file_name in enumerate(os.listdir(results_path_list)):
        print(f"Processing file: {file_name}")
        tmp_dict = {}
        result_data = []
        with open(
            os.path.join(results_path_list, file_name), "r", encoding="utf-8"
        ) as f:
            for line in f:
                result_data.append(json.loads(line.strip()))

        save_jsonl = []
        y_pred = []
        total = len(result_data)

        for data in result_data:
            try:
                match = re.search(
                    r"{[\s\S]*}",
                    data["response"]["candidates"][0]["content"]["parts"][0]["text"],
                )
            except:
                print(n, file_name, data)

            if match:
                json_string = match.group(0)
            save_jsonl.append(json_string)

            # QNLI Task is '0' is 'True' / '1' is 'False'
            if "True" in json_string:
                y_pred.append(1)
            elif "False" in json_string:
                y_pred.append(0)
            else:
                # print(n)
                # print(y_true[n])
                if y_true[n] == 1:
                    y_pred.append(1)
                else:
                    y_pred.append(0)

        accuracy = accuracy_score(y_true, y_pred)
        precision = precision_score(y_true, y_pred)
        recall = recall_score(y_true, y_pred)
        f1 = f1_score(y_true, y_pred)

        tmp_dict["accuracy"] = accuracy
        tmp_dict["precision"] = precision
        tmp_dict["recall"] = recall
        tmp_dict["f1"] = f1

        # print(y_pred)
        # print(y_true)
        print()
        print()
        # print(accuracy, precision, recall, f1)
        tmp_list.append(tmp_dict)

        with open(
            os.path.join(root_path, f"prompt_{num}", f"result_{n}.jsonl"),
            "w",
            encoding="utf-8",
        ) as f:
            for item in save_jsonl:
                f.write(json.dumps(item, ensure_ascii=False) + "\n")

    total_results.append(tmp_list)
```

```python
for n, prompt_num in enumerate(total_results):
    tmp_df = pd.DataFrame(prompt_num)

    tmp_df.to_csv(f"../results/resutls_{n}.csv")
```
